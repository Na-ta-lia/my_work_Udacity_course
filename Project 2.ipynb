```python
import time
import pandas as pd
import numpy as np

#I followed the examples in this course 
#and the help from the practice problems and solution 


#defining function that filters and presents the data

def load_filter_present_data(city, month, day):
    """this function loads the data, modifies the data by adding new variables, filters the information asked by the user and prepares the output that the user will read """

    CITY_DATA = { 'chicago': 'chicago.csv',
              'nyc': 'new_york_city.csv',
              'washington': 'washington.csv' }

    # load data file into a dataframe
    df = pd.read_csv(CITY_DATA[city])

    # convert the Start Time column to datetime
    df['Start Time'] = pd.to_datetime(df['Start Time'])

    # extract month and day of week from Start Time to create new columns
    df['month'] = df['Start Time'].dt.month
    df['day_of_week'] = df['Start Time'].dt.weekday_name

    #create column with start hour
    
    df['hour1'] = df['Start Time'].dt.hour
        
    df['start-end route']=df[['Start Station', 'End Station']].agg('-'.join, axis=1)
    
    # filter by month 
    if month != 'all':
        months = ['january', 'february', 'march', 'april', 'may', 'june']
        month = months.index(month) + 1

        df = df[df['month'] == month]

    # filter by day of week 
    if day != 'all':
        # filter by day of week
        df = df[df['day_of_week'] == day.title()]
             
    #prepare output              
    pop_month=df['month'].mode()[0]
    print(f"the most popular month of bikeshare is {pop_month}\n")
      
    pop_weekday=df['day_of_week'].mode()[0]
    print(f"the most popular day of the week is {pop_weekday}\n")
    
    pop_hour=df['hour1'].mode()[0]
    print(f"the most popular hour is {pop_hour} \n")
    
    start_station=df['Start Station'].mode()[0]
    print(f"the most popular start Station is {start_station} \n")
    
    end_station=df['End Station'].mode()[0]
    print(f"the most popular end Station is {end_station} \n")
    
    pop_route=df['start-end route'].mode()[0]
    print(f"the most popular route is {pop_route}\n ")
    
    total_time=df['Trip Duration'].sum()
    total_time=total_time/3600
    print(f"The total time travel is {total_time} hours\n")
 
    average_time=df['Trip Duration'].mean()
    print(f"the average travel time is {average_time} seconds\n ")
  
    user_type=df['User Type'].value_counts()
    print(f"the user types are {user_type}\n")
  
    if city == 'nyc' or city == 'chicago':
        user_gender=df['Gender'].value_counts()
        print(f"the user gender are {user_gender}\n")

        birth_year=df['Birth Year'].mode()[0]
        print(f"the most common birth year is {birth_year}\n")

    print("Now some general statistic information: \n")
    print(df.describe())
    
    #def display_data(view_data,df):
    #I could not find a way to export the data frame from the previous function and save it in the workspace so I can use it 
    #in the function display_data, so I decided to add this part to the first function:
    
    view_data=input("Do you want to see the first 5 rows of raw data?  ")
           
    if view_data == "yes":
        print("pruebo si lee hasta el yes")
        print(df.iloc[0:5])    

    i=0
    while view_data == "yes": 
        i=i+5
        j=i+5   
        
        view_data=input("Do you want to see the next 5 rows of raw data?  ")
        
        if view_data == "yes":

            print("pruebo si lee hasta el segundo yes")
            print(df.iloc[i:j])    

        else:
            break

            
    
#set of questions that assignes variables

"""here the function finishes and starts the code for the user interaction"""

answer=input("Hello, welcome to my set of questions that will guide you to analyze the bikeshare data. \n Are you ready to start?")

if answer=="OK":
    answer="yes"
if answer=="ok":
    answer="yes"
if answer=="ok":
    answer="yes"
if answer=="sure":
    answer="yes"
if answer=="of course":
    answer="yes"
if answer=="absolutely":
    answer="yes"
    
while answer!="yes":
    answer=input("...and now?") 
    if answer=="OK":
        answer="yes"
    if answer=="ok":
        answer="yes"
    if answer=="ok":
        answer="yes"
    if answer=="sure":
        answer="yes"
    if answer=="of course":
        answer="yes"
    if answer=="absolutely":
        answer="yes"

if answer=="yes":
    city=input("Please enter the name of the city (Chicago, NYC or Washington):\n")
    city = city.lower()
    if city=="new york":
        city="nyc"    
    if city=="new york city":
        city="nyc"    
    if city=="ny":
        city="nyc"    
        
    while city!="chicago" and city!="nyc" and city!="washington":
        city=input("Please, chose only between Chicago, NYC or Washington. Please enter city:")

        city = city.lower()
        if city=="new york":
            city="nyc"    
        if city=="new york city":
            city="nyc"    
        if city=="ny":
            city="nyc"    

print(f"filtering data for {city}...")

month=input("Please enter month. Please consider from January to June, (write \'all' if you want too consider all of them):\n")
month= month.lower()    

while month not in ['january', 'february', 'march', 'april', 'may', 'june','all']: 
    month=input("Please chose a month between January and June")

print(f"processing data for {month}...")

wday=input("please enter weekday (write \'all' if you want too consider all of them):\n")
print(f"processing data for {wday}...")

print(f"Loading data for {wday}, during the month of {month} in the city of {city}")

print(load_filter_present_data(city, month , wday))




    
```
